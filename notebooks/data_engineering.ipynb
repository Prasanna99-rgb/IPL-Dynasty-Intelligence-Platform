```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)
```

```python
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
players = pd.read_csv("players.csv")
seasons = pd.read_csv("seasons.csv")
```

```python
matches = matches.replace({
    "Delhi Daredevils": "Delhi Capitals",
    "Kings XI Punjab": "Punjab Kings",
    "Royal Challengers Bangalore": "Royal Challengers Bengaluru"
})

deliveries = deliveries.replace({
    "Delhi Daredevils": "Delhi Capitals",
    "Kings XI Punjab": "Punjab Kings",
    "Royal Challengers Bangalore": "Royal Challengers Bengaluru"
})
```

```python
print("MATCHES:", matches.shape)
print("DELIVERIES:", deliveries.shape)
print("PLAYERS:", players.shape)
print("SEASONS:", seasons.shape)
```

```python
print(matches.columns.tolist())
print(deliveries.columns.tolist())
print(players.columns.tolist())
print(seasons.columns.tolist())
```

```python
for name, df in {
    "matches": matches,
    "deliveries": deliveries,
    "players": players,
    "seasons": seasons
}.items():

    print(f"\n{name.upper()}")
    print(df.isnull().sum())
```

```python
matches = matches.drop_duplicates()
deliveries = deliveries.drop_duplicates()
players = players.drop_duplicates()
seasons = seasons.drop_duplicates()
```

```python
matches.columns = matches.columns.str.lower().str.strip()
deliveries.columns = deliveries.columns.str.lower().str.strip()
players.columns = players.columns.str.lower().str.strip()
seasons.columns = seasons.columns.str.lower().str.strip()
```

```python
all_teams = sorted(
    set(matches["team1"].dropna().unique())
    |
    set(matches["team2"].dropna().unique())
)

print("Total Teams:", len(all_teams))
print(all_teams)
```

```python
match_runs = (
    deliveries
    .groupby("match_id")["total_runs"]
    .sum()
    .reset_index()
)

match_runs.columns = [
    "match_id",
    "match_total_runs"
]

match_runs.head()
```

```python
matches = matches.merge(
    match_runs,
    on="match_id",
    how="left"
)

matches.head()
```

```python
matches["match_status"] = np.where(
    matches["winner"].notnull(),
    "Completed",
    "No Result"
)
```

```python
team_stats = (
    matches.groupby("winner")
    .agg(
        matches_won=("winner", "count")
    )
    .reset_index()
)

team_stats = team_stats.sort_values(
    "matches_won",
    ascending=False
)

team_stats.head()
```

```python
venue_stats = (
    matches.groupby("venue")
    .agg(
        total_matches=("match_id", "count"),
        avg_runs=("match_total_runs", "mean")
    )
    .reset_index()
)

venue_stats.head()
```

```python
ball_by_ball = deliveries.merge(
    matches[
        [
            "match_id",
            "season",
            "venue",
            "winner"
        ]
    ],
    on="match_id",
    how="left"
)

ball_by_ball.head()
```

```python
ball_by_ball["boundary_type"] = np.where(
    ball_by_ball["batsman_runs"].isin([4, 6]),
    "Boundary",
    "Non-Boundary"
)
```

```python
ball_by_ball["run_type"] = np.where(
    ball_by_ball["batsman_runs"] == 4,
    "Four",
    "Other"
)
```

```python
ball_by_ball["six_flag"] = np.where(
    ball_by_ball["batsman_runs"] == 6,
    "Six",
    "Other"
)
```

```python
ball_by_ball["ball_result"] = np.where(
    ball_by_ball["total_runs"] == 0,
    "Dot Ball",
    "Scoring Ball"
)
```

```python
ball_by_ball["wicket_status"] = np.where(
    ball_by_ball["is_wicket"] == 1,
    "Wicket",
    "Not Out"
)
```

```python
ball_by_ball["extra_status"] = np.where(
    ball_by_ball["extra_runs"] > 0,
    "Extra",
    "No Extra"
)
```

```python
team_matches = pd.concat([
    matches[["match_id", "team1"]]
        .rename(columns={"team1": "team"}),

    matches[["match_id", "team2"]]
        .rename(columns={"team2": "team"})
])

team_matches.head()
```

```python
team_match_count = (
    team_matches.groupby("team")
    .size()
    .reset_index(name="matches_played")
)

team_match_count.head()
```

```python
team_stats = team_stats.merge(
    team_match_count,
    left_on="winner",
    right_on="team",
    how="left"
)

team_stats["win_percentage"] = round(
    (
        team_stats["matches_won"]
        /
        team_stats["matches_played"]
    ) * 100,
    2
)

team_stats.head()
```

```python
matches.to_csv(
    "matches_clean.csv",
    index=False
)

ball_by_ball.to_csv(
    "ball_by_ball_clean.csv",
    index=False
)

team_stats.to_csv(
    "team_stats.csv",
    index=False
)

venue_stats.to_csv(
    "venue_stats.csv",
    index=False
)
```
